## Research Question

Do air pollutants predict the severity (length/cost) of a power outage, and do these predictors change by region?

## Background and Prior Work

&emsp;In the United States, air quality is monitored and defined by the Environmental Protection Agency (EPA) nested in the government’s executive branch. Air quality is characterized as the concentration of specific pollutants in the atmosphere and their resulting effects on the health of the public/environment. The EPA’s Air Quality Index (AQI) ranges from “Good” to “Hazardous” where the severity of pollutant diffusion across the atmosphere poses little risk to the public (“Good”) to everyone being increasingly likely to experience adverse effects from its exposure (“Hazardous”). The AQI’s index values intuitively rate air quality by having a higher AQI value meaning the greater the level of air pollution and vice versa.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) It is important to note however that the AQI is a composite measure and takes multiple air pollutants into account rather than just an individual pollutant concentration. 

&emsp; As the relationship between power outages and air pollution is concerned, previous work has been done in Ecuador, specifically in the capital, Quito, and their 2023 and 2024 power outages which saw large scale blackouts for several hours for consecutive months. Researchers Vallejo and their colleagues posit that Quito’s blackouts, “...revealed a significant deterioration in atmospheric pollution, primarily driven by the widespread use of diesel generators, increased vehicular traffic, and adverse meteorological conditions…”<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Vallejo and colleagues’ work in the context of Ecuador’s 2023 and 2024 power outages reveals an interesting interaction between infrastructural issues such as blackouts and the increased usage of alternative power sources which produce pollutants and deteriorate the atmosphere. Another work that explores the relationship between air pollution and power outages takes a similar look at how Delhi, India and its citizens grapple with power outages through reliance on backup solutions like diesel generators. Yatang Lin and Dana Kassem note that diesel generators “...emit 3.5 times more nitrogen oxides and two times more particulate matter than coal-fired plants for the same amount of energy,” which is worsened by the fact that Delhi’s power outages are locally combatted through “The frequent on-off cycling of backup generators…” which “...further increases fuel consumption and pollution,”<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Lin and Kassem’s findings further display the downstream pollutive effects of power outages and how they influence worsening air quality. To be subject to frequent blackouts and power outages coupled with a reliance on alternative power sources which negatively impact air quality establishes this relationship wherein power outages can possibly predict worsening air quality/pollution. 

&emsp; An important consideration when discussing air quality involves the health complications that come with exposure to air pollutants. As detailed by California’s Air Resources Board, the reliance on backup generators (which are primarily diesel generators) and their emission of diesel particulate matter (DPM) is considered a toxic air contaminant implicated in the development of cancer and increasing the lung’s susceptibility to damage<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) The ability for air pollution and power outages to influence one another has an impact on health either way through exposure to harmful airborne substances or as how Zhang and colleagues put it, power outages threaten vulnerable populations as they suggest an increased risk of COPD exacerbation, as a result of their dependency on electricity<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) The importance of air quality and power reliability influences on the health of the public cannot be understated.

&emsp; This project examines the relationship between air pollution and power outages within the context of the United States and intends to explore the inverse of the previously established case studies done in Quito and Delhi. We intend to examine if air pollutants are a predictor for the severity (length/cost) of power outages nationally, and if these predictors (specific air pollutants) are subject to variability per US region.

The ideal dataset for this question would measure multiple air pollutants at fine geographic resolution (county level or better), annually, over at least a decade, across all U.S. states — and a corresponding dataset recording power outage events with location (at the same level as the air pollutant dataset), duration, customers affected, and cause classification over the same time period. Ideally both datasets would share a common geographic identifier (such as county FIPS code or zip code) to enable direct merging without aggregation loss. Storage in a flat CSV or Excel format would make standard data science workflows in Python/pandas easy.

CACES U.S. Air Quality Data - https://www.caces.us/data, requires completing a brief access form. 
This dataset provides annual, county-level, population-weighted ambient concentration estimates for ten pollutants: CO, NO₂, O₃, BC, PM10, PM2.5, SO₂, COA, HOA, and PNC, covering 1979–2020. Key columns are `fips` (county identifier), `pollutant`, `year`, `pred_wght` (the population-weighted concentration estimate), `state_abbr`, `lat`, and `lon`. Estimates are derived from EPA monitoring networks combined with land-use regression models, providing coverage even in areas without monitoring stations. We could aggregate to annual state means using `state_abbr` and `year` as keys to match the geographic resolution of the outage data. 

Purdue Major Power Outage Risks in the U.S. - https://engineering.purdue.edu/LASCI/research-data/outages, direct download, no application required. 
This dataset records 1,534 major power outage events in the contiguous U.S. from January 2000 to July 2016, with one row per outage event. Key variables include `U.S._STATE` and `POSTAL.CODE` (state-level location), `OUTAGE.DURATION` (minutes), `CUSTOMERS.AFFECTED`, `DEMAND.LOSS.MW`, `CAUSE.CATEGORY` (e.g., severe weather, intentional attack, equipment failure), `CLIMATE.REGION` (nine U.S. climate zones), and sector-level electricity prices (`RES.PRICE`, `COM.PRICE`, `IND.PRICE`, `TOTAL.PRICE`). Because location is recorded only at the state level, we aggregate outage events to annual state means and counts before merging with CACES, yielding a final merged dataset of state × year observations covering 2000–2016 across up to 48 states. It would mean the loss of more precise location data using longitude/latitude, but the overall power outage data itself is useful
